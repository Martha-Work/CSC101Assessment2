```python
# Simple program to read a text file and show basic info

filename = "my_data.txt"   # the file we want to read

# Read all lines from the file
with open(filename, "r") as file:
    lines = file.readlines()

# Print the file contents
print("=== MY DATA STORY ===\n")
for line in lines:
    print(line.strip())   # remove extra spaces/newlines

# Basic statistics
num_lines = len(lines)                           # how many lines
num_words = sum(len(line.split()) for line in lines)  # total words
longest_line = max(lines, key=len).strip()       # longest line
total_chars = sum(len(line) for line in lines)   # total characters

# Show the results
print("\n--- FILE STATISTICS ---")
print("Lines:", num_lines)
print("Words:", num_words)
print("Longest line:", longest_line)
print("Characters:", total_chars)
```
