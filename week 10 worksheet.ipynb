1.1 Problem 1: Tip Calculator
1.1.1 The Scenario
You work at a Perth restaurant. The manager wants a simple tool where staff can enter:

The bill amount (e.g., $45.80)
The tip percentage (e.g., 15%)
And get back:

The tip amount
The total amount to charge
1.1.2 What You Must Submit
STEP 1: Understand the Problem

In your own words, what is this program supposed to do? (2-3 sentences)

[This program is meant to help restaurant staff work out how much tip to add to a customer’s bill. The staff member enters the bill amount and the tip percentage they want, and the program calculates the exact tip amount. It then adds the tip to the bill and shows the final total the customer needs to pay.]

STEP 2: Identify Inputs and Outputs

INPUTS:

[The bill amount and tip percentage]
[Number($45.80, 15% tip)]
OUTPUTS:

[The total amount to charge(bill+tip)]
[Currency amount]
STEP 3: Work It By Hand

Show me you can do this with a calculator. Use this example:

Bill amount: $45.80
Tip percentage: 15%
Write the steps:

[Convert 15% to decimal =0.15]
[Multiply the bill by tip rate to find the tip 45.80*0.15=/ 6.87]
[Add the tip to the bill to get the total 45.80+6.87= /52.67]
[Total amount to charge is /52.67]
1.2 Problem 2: Late Fee Calculator
1.2.1 The Scenario
You work for a library in Perth. Books are due back in 14 days. For every day late, there’s a /0.50 fine (maximum /10.00).

The librarian wants a program where they enter:

How many days the book is overdue
And get back:

The fine amount (or “Maximum fine reached” if it’s over /10)
1.2.2 What You Must Submit
STEP 1: Understand the Problem

In your own words, what is this program supposed to do? (2-3 sentences)

[This program is meant to help the librarian to find how much a customer owes when a book is returned late. The librarian enters how many days overdue the book is, and the program calculates the fine based on /0.50 per day. If the fine goes above $10, the program should show that the maximum fine has been reached.]

STEP 2: Identify Inputs and Outputs

INPUTS:

[Days overdue]
OUTPUTS:

[The fine amount based on how many days the book is overdue]
[If the fine is more than /10 the program should stop and show the maximum fine. If the book is 0 days late, the fine should be $0]
STEP 3: Work It By Hand

Show me you understand the logic. Calculate these three cases manually:

Case A: Book is 5 days late

My calculation: [5*/0.50]
Fine: [$2.50]
Case B: Book is 25 days late

My calculation: [25*/0.50]
Fine: [$12.50]
Special note: [the fine is higher than limit so the program must stop from going above the maximum allowed amount]
Case C: Book is 0 days late

Fine: [No days overdue- no fine]
