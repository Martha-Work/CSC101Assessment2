```python
def load_transactions(filename):
    """
    Read lines from a transaction file safely.

    Returns:
        A list of cleaned lines (strings) with blank lines removed.
        If the file does not exist, returns an empty list.
    """

    try:
        # Try to open the file. If the file does NOT exist,
        # Python will automatically jump to the except block.
        with open(filename, 'r') as file:
            lines = file.readlines()

        # Clean the lines:
        # - strip() removes spaces and newline characters
        # - the 'if line.strip()' part removes empty lines
        cleaned_lines = [line.strip() for line in lines if line.strip()]

        # Return the cleaned list of transaction lines
        return cleaned_lines

    except FileNotFoundError:
        # If the file cannot be found, show a friendly message
        # instead of crashing the whole program.
        print(f"File '{filename}' not found!")
        return []
```

```python
def process_transactions(lines):
    """Basic structure for turning lines into dictionaries."""
    transactions = []

    for line in lines:
        try:
            description, amount_str = line.split(",")
            amount = float(amount_str)

            transaction = {
                "description": description,
                "amount": amount
            }

            transactions.append(transaction)

        except Exception:
            # Skip corrupted lines
            continue

    return transactions
```

```python
def process_transactions(lines):
    transactions = []
    for line in lines:
        line = line.strip()

        # Skip empty lines
        if not line:
            continue

        # Skip lines without a comma
        if "," not in line:
            continue

        description, amount_str = line.split(",", 1)

        # Try converting amount to float
        try:
            amount = float(amount_str)
        except ValueError:
            # Non-numeric amount → skip
            continue

        transactions.append({
            "description": description,
            "amount": amount
        })

    return transactions
```

```python
def process_transactions(lines):
    transactions = []
    failed = 0

    for line in lines:
        line = line.strip()

        if not line:
            failed += 1
            continue

        if "," not in line:
            failed += 1
            continue

        description, amount_str = line.split(",", 1)

        try:
            amount = float(amount_str)
        except ValueError:
            failed += 1
            continue

        transactions.append({
            "description": description,
            "amount": amount
        })

    return transactions, failed
```

```python
def summarise_transactions(transactions):
    """
    Take a list of transaction dictionaries and calculate:
    - total number of transactions
    - total income (positive amounts)
    - total expenses (negative amounts)
    - net amount (income + expenses)
    """

    # Count how many valid transactions there are
    total_transactions = len(transactions)

    # Add up all positive amounts
    total_income = sum(t["amount"] for t in transactions if t["amount"] > 0)

    # Add up all negative amounts
    total_expenses = sum(t["amount"] for t in transactions if t["amount"] < 0)

    # Net amount is income + expenses
    net_amount = total_income + total_expenses

    # Return results in a dictionary
    return {
        "total_transactions": total_transactions,
        "total_income": total_income,
        "total_expenses": total_expenses,
        "net_amount": net_amount
    }
```

```python
def display_summary(summary):
    """
    Display the results of the transaction summary in a clear format.
    Expects a dictionary with:
    - total_transactions
    - total_income
    - total_expenses
    - net_amount
    """

    print("=== Transaction Summary ===")
    print(f"Total transactions processed: {summary['total_transactions']}")
    print(f"Total income: ${summary['total_income']:.2f}")
    print(f"Total expenses: ${summary['total_expenses']:.2f}")
    print(f"Net amount: ${summary['net_amount']:.2f}")
```
