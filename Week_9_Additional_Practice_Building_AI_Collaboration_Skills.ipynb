Using the sales_data.csv file, analyse regional performance.
Please write Python code that:

1. Calculates total sales by Region  
2. Calculates average sale amount by Region  
3. Counts number of transactions per Region  
4. Returns a clean, manager‑friendly summary table sorted from highest‑performing region to lowest  

The dataset includes columns: Date, Product, Quantity, Unit_Price (with $ signs), Total_Sale, Sales_Rep, and Region.  
Make sure to clean Unit_Price if needed and use Total_Sale for revenue calculations.  
Format currency values clearly (e.g., $12,345.67) and include a short explanation of the results suitable for a business presentation.


```python
# Sales Performance by Region
import pandas as pd

# Load the sales data
sales_df = pd.read_csv('../data/sales_data.csv')

# Clean Unit_Price column (remove $ and convert to float)
sales_df['Unit_Price'] = (
    sales_df['Unit_Price']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

# Ensure Total_Sale is numeric (in case formatting issues exist)
sales_df['Total_Sale'] = (
    sales_df['Total_Sale']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

# --- REGIONAL PERFORMANCE CALCULATIONS ---

# Total revenue by region
total_sales = sales_df.groupby('Region')['Total_Sale'].sum()

# Average sale amount by region
avg_sales = sales_df.groupby('Region')['Total_Sale'].mean()

# Number of transactions by region
transaction_counts = sales_df.groupby('Region')['Total_Sale'].count()

# Combine into one summary table
summary_df = pd.DataFrame({
    'Total_Sales': total_sales,
    'Average_Sale': avg_sales,
    'Transaction_Count': transaction_counts
})

# Sort by highest total sales
summary_df = summary_df.sort_values(by='Total_Sales', ascending=False)

# Format currency for readability
summary_df['Total_Sales'] = summary_df['Total_Sales'].map('${:,.2f}'.format)
summary_df['Average_Sale'] = summary_df['Average_Sale'].map('${:,.2f}'.format)

# Display results
print("\n📊 Regional Sales Performance Summary")
print("=" * 50)
print(summary_df)
```

```python
# Top performing Products
import pandas as pd

# Load the sales data
sales_df = pd.read_csv('../data/sales_data.csv')

# CLEANING

# Clean Unit_Price column
sales_df['Unit_Price'] = (
    sales_df['Unit_Price']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

# Clean Total_Sale column
sales_df['Total_Sale'] = (
    sales_df['Total_Sale']
    .replace('[\$,]', '', regex=True)
    .astype(float)
)

# TOP PRODUCTS BY REVENUE

top_revenue = (
    sales_df.groupby('Product')['Total_Sale']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

top_revenue['Total_Sale'] = top_revenue['Total_Sale'].map('${:,.2f}'.format)

# TOP PRODUCTS BY UNITS SOLD

top_units = (
    sales_df.groupby('Product')['Quantity']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

# PROFIT MARGIN POTENTIAL (using average unit price)

profit_margin = (
    sales_df.groupby('Product')['Unit_Price']
    .mean()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

profit_margin['Unit_Price'] = profit_margin['Unit_Price'].map('${:,.2f}'.format)

# --- DISPLAY RESULTS ---

print("\n🏆 Top 5 Products by Total Revenue")
print("=" * 50)
print(top_revenue)

print("\n📦 Top 5 Products by Units Sold")
print("=" * 50)
print(top_units)

print("\n💰 Products with Highest Profit Margin Potential (Avg Unit Price)")
print("=" * 50)
print(profit_margin)

# MANAGER SUMMARY

print("\n📊 Manager Summary")
print("=" * 50)
print(
    "The top‑performing products show strong alignment between revenue and unit sales. "
    "High‑priced items with strong sales volume should be prioritised for next quarter’s inventory. "
    "Products with high average unit prices indicate strong margin potential and may justify increased stock levels."
)
```

```python
# Spending Trend Analysis
import pandas as pd

# Load transaction data
df = pd.read_csv('../data/transactions.csv')

# 1. Convert dates to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Extract useful time features
df['DayOfWeek'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.to_period('M')

# 2. Spending trends by day of week
weekday_spending = (
    df.groupby('DayOfWeek')['Amount']
    .sum()
    .sort_values(ascending=False)
)

print("=== Spending by Day of Week ===")
print(weekday_spending)

# 3. Is spending increasing over time?
monthly_spending = (
    df.groupby('Month')['Amount']
    .sum()
    .sort_index()
)

print("\n=== Monthly Spending Trend ===")
print(monthly_spending)

# Simple trend check
if monthly_spending.iloc[-1] > monthly_spending.iloc[0]:
    print("\nTrend: Spending has increased over time.")
else:
    print("\nTrend: Spending has decreased or stayed stable over time.")

# 4. Detect unusual spending patterns (outliers)
mean_amt = df['Amount'].mean()
std_amt = df['Amount'].std()

# Flag transactions more than 2 standard deviations from mean
df['Unusual'] = df['Amount'] > (mean_amt + 2 * std_amt)

unusual_spending = df[df['Unusual']]

print("\n=== Unusual Spending Transactions ===")
print(unusual_spending[['Date', 'Amount', 'Description']])
```

```python
# Budget Category Analysis
import pandas as pd

# Load data
df = pd.read_csv('../data/transactions.csv')

# Convert dates
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.to_period('M')

# 1. Average monthly spending by category
monthly = df.groupby(['Month', 'Category'])['Amount'].sum().reset_index()

avg_monthly = (
    monthly.groupby('Category')['Amount']
    .mean()
    .sort_values(ascending=False)
)

print("=== Average Monthly Spending by Category ===")
print(avg_monthly)

# 2. Categories with high variation
variation = (
    monthly.groupby('Category')['Amount']
    .std()
    .sort_values(ascending=False)
)

print("\n=== Spending Variation (Standard Deviation) ===")
print(variation)

# 3. Suggested realistic budget amounts
suggested_budget = avg_monthly + variation.fillna(0)

print("\n=== Suggested Budget Amounts (Mean + Variation) ===")
print(suggested_budget)

# 4. Categories where spending might be reduced

max_monthly = (
    monthly.groupby('Category')['Amount']
    .max()
)

overspend_gap = max_monthly - avg_monthly
overspend_gap = overspend_gap.sort_values(ascending=False)

print("\n=== Categories With Potential to Reduce Spending ===")
print(overspend_gap)
```

```python
import pandas as pd

def finance_coach(df):
    # 1. Calculate category percentages
    total = df['Amount'].sum()
    category_percent = (df.groupby('Category')['Amount'].sum() / total) * 100

    # Typical recommended spending percentages (example values)
    recommended = {
        'Food': 15,
        'Transport': 10,
        'Entertainment': 5,
        'Shopping': 10,
        'Bills': 25,
        'Health': 5,
        'Other': 5
    }

    advice = []

    # 2. Identify overspending + suggest savings
    for category, pct in category_percent.items():
        rec = recommended.get(category, 10)  # default recommendation

        if pct > rec:
            advice.append(
                f"⚠️ You spent {pct:.1f}% on {category}, which is above the typical {rec}%.\n"
                f"💡 Consider reducing spending in {category} to save money."
            )
        else:
            advice.append(
                f"✅ Great job! Your spending on {category} is within a healthy range ({pct:.1f}%)."
            )

    # 3. Format advice nicely
    final_message = "\n\n".join(advice)
    return final_message


# Example usage:
# print(finance_coach(df))
```
